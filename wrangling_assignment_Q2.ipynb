```python
# Q2, part 2
df = pd.read_csv('./sharks.csv', low_memory=False) # load in shark data
print(df.shape) # list dimensions of dataframe(rows, columns)
df.head() # show first 5 rows
```

```python
type = df['Type'] # creates variable called type set equal to array of type column values
type.unique() # prints values of type - objects
# many of these value types could be grouped together since they mean the same thing
```

```python
df['Type'].value_counts()
```

```python
type = type.replace(['Watercraft', 'Boat', 'Boating', 'Boatomg'], 'Sea Disaster')
type.value_counts()
# combines all boating and water craft incidents under sea disasters since they all mean the same thing
```

```python
# all the invalid, questionable, unconfirmed, etc incidents can be represented uniformly as missing values
type = type.replace(['Invalid', 'Questionable','Unconfirmed','Unverified','Under investigation'],np.nan) # removed missing values
type.value_counts()
```

```python
# replaces the type column in shark data with the cleaned version that is saved under the temporary vector, type
df['Type'] = type
del type

df['Type'].value_counts()
```

```python
# Q2, part 3
url = 'http://www.vcsc.virginia.gov/pretrialdataproject/October%202017%20Cohort_Virginia%20Pretrial%20Data%20Project_Deidentified%20FINAL%20Update_10272021.csv'
df = pd.read_csv(url,low_memory=False) # downloads and loads the .csv file from url
print(df.shape) # list dimensions of dataframe(rows, columns)
df.head() # show first 5 rows
```

```python
PTrelease = df['WhetherDefendantWasReleasedPretrial']
print(PTrelease.unique())
# 9 = unclear
```

```python
sum(PTrelease.isnull()) # total missing values
```

```python
# change unlcear values to missing values
PTrelease = PTrelease.replace(9,np.nan)
print(PTrelease.value_counts())
df['WhetherDefendantWasReleasedPretrial'] = PTrelease # Replace data column with cleaned values
del PTrelease
```

```python
# Q2, part 4
imposed = df["ImposedSentenceAllChargeInContactEvent"]
print(imposed.shape)
print(imposed.unique())
imposed.head()
```

```python
imposed = pd.to_numeric(imposed,errors='coerce') # coerce to numeric values
imposed_null = imposed.isnull() # creates a missing dummy variable
print(np.sum(imposed_null)) # finds number of missing values
```

```python
type = df['SentenceTypeAllChargesAtConvictionInContactEvent']
```

```python
print(pd.crosstab(imposed_null, type)) # cross tab into bins by type
```

```python
imposed = imposed.mask(type == 4, 0) # replace category 4 with 0 (false) because 4 means charges were dismissed
imposed = imposed.mask(type == 9, np.nan) # replace category 9 with nan

imposed_null = imposed.isnull() # create a new missing dummy with modified categories
print(pd.crosstab(imposed_null, type), '\n') # \n for new line
print(np.sum(imposed_null)) # see new number of null values is better

df['ImposedSentenceAllChargeInContactEvent'] = imposed # replace data with cleaned version
del imposed, type # delete temporary length/type variables
```
